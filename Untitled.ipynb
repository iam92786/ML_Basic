```python
#Regression : The term regression is used when you try to find the relationship between variables.
#That relationship is used to predict the outcome of future events.

#Linear Regression : Linear regression uses the relationship between the data-points to draw a straight line through all them.
#This line can be used to predict future values.
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)
#Execute a method that returns some important key values of Linear Regression:

#Create a function that uses the slope and intercept values to return a new value.
def myfunction(x):
    return slope * x + intercept
    #return valu repersent y-axis the corresponding x value will be placed:

mymodel = list(map(myfunction, x)) 
#Run each value of the x array through the function. This will result in a new array with new values for the y-axis.

plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunction(x):
    return slope * x + intercept

speed = myfunction(10)

print(speed)
```

```python
#Polynomial Regression: If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.
#like linear regression, poly regression also uses the relationship between the variables x and y to find the best way to draw a line through the data points.
```

```python
#In the example below, we have registered 18 cars as they were passing a certain tollbooth.
#We have registered the car's speed, and the time of day (hour) the passing occurred.
#The x-axis represents the hours of the day and the y-axis represents the speed:
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x,y,3))
#NumPy has a method that lets us make a polynomial model.

myline = np.linspace(1,22,100)

plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```
